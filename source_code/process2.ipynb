```python
#Tatum Miller Process Control Assignment 1


import numpy as np

#define ODE
def diffeq1(y, t, b, c):
    theta, omega = y
    dydt = [omega, -b*omega - c*np.sin(theta)]
    return dydt

#define constants
b= 0.25
c = 5.0

#define intial conditons
y0 = [np.pi - 0.1, 0.0]

#define time
t = np.linspace(0, 10, 101)

#solve
from scipy.integrate import odeint

sol = odeint(diffeq1, y0, t, args=(b, c))

#plot
import matplotlib.pyplot as plt
plt.plot(t, sol[:, 0], 'b', label='theta(t)')
plt.plot(t, sol[:, 1], 'g', label='omega(t)')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```
